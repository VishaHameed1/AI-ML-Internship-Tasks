# TASK 5 - STEP 5: Load Pre-trained DistilGPT2 Model


```python
# TASK 5 - STEP 5: Load Pre-trained DistilGPT2 Model

print("\n" + "="*50)
print("LOADING MODEL")
print("="*50)

model_name = "distilgpt2"

print(f"📥 Loading {model_name}...")
print("⏳ First time download may take 2-3 minutes...")

try:
    from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
    
    # Load tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)
    
    # Add padding token (required)
    tokenizer.pad_token = tokenizer.eos_token
    
    # Create text generation pipeline
    generator = pipeline(
        'text-generation',
        model=model,
        tokenizer=tokenizer,
        device=-1,  # -1 = CPU, 0 = GPU if available
        max_length=100,
        temperature=0.8,
        do_sample=True
    )
    
    print("✅ Model loaded successfully!")
    print(f"📊 Model size: {model.num_parameters():,} parameters")
    print("💡 Model is running on CPU")
    
except Exception as e:
    print(f"⚠️ Error loading model: {e}")
    generator = None
    print("📌 Will use fallback responses mode")

print("\n✅ STEP 5 Complete!")
```

```python
# TASK 5 - STEP 7 (FIXED): Simplified Response Generator

print("\n" + "="*50)
print("BUILDING SIMPLE RESPONSE GENERATOR")
print("="*50)

def get_emotion_simple(text):
    """Detect emotion from user input"""
    text_lower = text.lower()
    
    if any(word in text_lower for word in ["stress", "overwhelm", "pressure"]):
        return "stress"
    elif any(word in text_lower for word in ["anxiety", "anxious", "worry", "nervous", "panic", "scared"]):
        return "anxiety"
    elif any(word in text_lower for word in ["sad", "depressed", "unhappy", "down"]):
        return "sad"
    elif any(word in text_lower for word in ["lonely", "alone", "isolated"]):
        return "lonely"
    elif any(word in text_lower for word in ["tired", "exhausted", "fatigue", "drained"]):
        return "tired"
    elif any(word in text_lower for word in ["angry", "mad", "frustrated"]):
        return "angry"
    elif any(word in text_lower for word in ["hopeless", "useless", "worthless"]):
        return "hopeless"
    elif any(word in text_lower for word in ["thank", "thanks", "appreciate"]):
        return "thank"
    else:
        return "general"

def get_response_simple(user_input):
    """Generate empathetic response without AI"""
    
    # Emergency check
    emergency_words = ["suicide", "kill myself", "want to die", "hurt myself", "self harm"]
    for word in emergency_words:
        if word in user_input.lower():
            return "⚠️ Please call iCall: +91-9152987821 (India) or 988 (USA). You matter! 💙"
    
    # Detect emotion
    emotion = get_emotion_simple(user_input)
    
    # Response database
    responses = {
        "stress": [
            "I hear you. Stress can feel overwhelming, but you're not alone. Take a deep breath with me. 🌬️",
            "It's completely normal to feel stressed. I'm here to listen. 💙",
            "Let's take a moment. You're stronger than you think."
        ],
        "anxiety": [
            "Anxiety is really tough. Let's try a grounding exercise - name 3 things you can see right now.",
            "I'm here with you. Take 5 deep breaths with me. This feeling will pass.",
            "It's okay to feel anxious. You're safe right now. 💙"
        ],
        "sad": [
            "I'm sorry you're feeling this way. Your feelings are valid. Would you like to talk about it? 💙",
            "Sometimes we all feel down. Better days will come. I'm here for you.",
            "Thank you for sharing. That takes courage. You matter."
        ],
        "lonely": [
            "You're not alone - I'm here talking with you. Even small connections matter. 💫",
            "Loneliness is hard. Is there someone you could reach out to?",
            "I see you and I hear you. You deserve connection and care."
        ],
        "tired": [
            "Make sure you're resting. Your mind and body need breaks to recharge. 🛌",
            "It's okay to take a break. Self-care isn't selfish.",
            "Exhaustion is real. Prioritize one small thing that gives you energy today."
        ],
        "angry": [
            "Anger is a natural emotion. Can we find a healthy way to release it? Maybe a walk?",
            "It's okay to be angry. Let's take a moment before responding.",
            "Your feelings are valid. Want to talk about what's bothering you?"
        ],
        "hopeless": [
            "I'm really glad you reached out. Please know that help is available. 💙",
            "These feelings are heavy. You don't have to carry them alone.",
            "There is hope, even when it's hard to see. You matter."
        ],
        "thank": [
            "You're very welcome! Taking care of your mental health is important. 💙",
            "I'm glad I could help. Remember, you're doing great!",
            "Thank you for being here. Keep taking care of yourself."
        ],
        "general": [
            "I'm here to listen. Tell me more about how you're feeling. 💙",
            "Thank you for sharing. That takes courage. I'm here for you.",
            "Your feelings are valid. Would you like to talk more about it?"
        ]
    }
    
    import random
    return random.choice(responses.get(emotion, responses["general"]))

def self_care_tip_simple():
    """Random self-care tip"""
    tips = [
        "🌬️ Take 3 deep breaths - inhale for 4, hold for 4, exhale for 4",
        "💧 Drink a glass of water right now",
        "🚶 Stand up and stretch your body for 1 minute",
        "📝 Write down one thing you're grateful for today",
        "🎵 Listen to your favorite calming song",
        "😴 Try to get 7-8 hours of sleep tonight",
        "📵 Take a 15-minute break from screens",
        "💬 Text or call someone you care about",
        "🌿 Spend 5 minutes outside or near a window"
    ]
    import random
    return random.choice(tips)

print("✅ Simple response generator ready!")
print("✅ Self-care tips loaded!")
print("\n✅ STEP 7 (FIXED) Complete!")
```
